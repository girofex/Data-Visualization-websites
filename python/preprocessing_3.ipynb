```python
import pandas as pd
import numpy as np
```

## Import CSV

```python
northamerica = pd.read_csv("../resources/og/US-and-Canada_aggregated_data_up_to-2025-10-11_0.csv")
northamerica.head()
```

```python
# I create a new column with the year
northamerica["YEAR"] = northamerica["WEEK"].str.slice(6)
northamerica.head()
```

```python
# I select only the protests
northamerica_cleaned = northamerica[northamerica["EVENT_TYPE"] == "Protests"]
```

```python
# I group by the year and sum the number of events
northamerica_cleaned = northamerica_cleaned.drop(columns=["FATALITIES", "ID", "CENTROID_LATITUDE", "CENTROID_LONGITUDE"])
northamerica_cleaned = northamerica_cleaned.groupby("YEAR").sum("EVENTS")
northamerica_cleaned.head()
```

```python
# Create the dataframe and save it as CSV
northamerica_cleaned.to_csv("../resources/plots/sectionthree/timeline_data.csv")
```
