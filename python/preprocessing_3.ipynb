```python
import pandas as pd
import numpy as np
```

### Define cleaning function

```python
def cleaning(df):
    df["YEAR"] = df["WEEK"].str.slice(6)
    df["YEAR"] = pd.to_numeric(df["YEAR"], errors="coerce")
    df = df[df["YEAR"] > 2018]
    df = df[df["EVENT_TYPE"] == "Protests"]
    df = df.drop(columns=["FATALITIES", "ID", "CENTROID_LATITUDE", "CENTROID_LONGITUDE"])
    df = df.groupby("YEAR").sum(numeric_only=True)
    return df
```

```python
africa_time = pd.read_csv("../resources/og/Africa_aggregated_data_up_to-2025-10-18.csv")
asiapacific_time = pd.read_csv("../resources/og/Asia-Pacific_aggregated_data_up_to-2025-10-11_0.csv")
europecentralasia_time = pd.read_csv("../resources/og/Europe-Central-Asia_aggregated_data_up_to-2025-10-11.csv")
latinamericacaribbean_time = pd.read_csv("../resources/og/Latin-America-the-Caribbean_aggregated_data_up_to-2025-10-18.csv")
middleeast_time = pd.read_csv("../resources/og/Middle-East_aggregated_data_up_to-2025-10-18.csv")
uscanada_time = pd.read_csv("../resources/og/US-and-Canada_aggregated_data_up_to-2025-10-11_0.csv")
```

```python
africa_time = cleaning(africa_time)
asiapacific_time = cleaning(asiapacific_time)
europecentralasia_time = cleaning(europecentralasia_time)
latinamericacaribbean_time = cleaning(latinamericacaribbean_time)
middleeast_time = cleaning(middleeast_time)
uscanada_time = cleaning(uscanada_time)
```
